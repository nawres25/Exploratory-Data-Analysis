Goal : As a business manager, try to find out the weak areas where you can work to make more profit.

IMPORTING LIBRARIES 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

DATA ANALYSIS 

```python
data = pd.read_csv("SampleSuperstore.csv")
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.columns
```

```python
# We can drop the country column because it's redundant since we have only the United States
data.drop(['Country'],axis=1)
```

Effect of Ship Mode and Segment on Profit

```python
fig = plt.figure(figsize=(8,5))
ship=data.groupby('Ship Mode').mean().reset_index()
plt.bar(ship['Ship Mode'],ship['Profit'],width=0.5)
plt.title('Profit for each ship mode')
plt.show()
```

```python
ship
```

```python
fig = plt.figure(figsize=(8,5))
ship=data.groupby('Segment').mean().reset_index()
plt.bar(ship['Segment'],ship['Profit'],width=0.5)
plt.title('Profit for each Segment')
plt.show()
```

```python
ship
```

The results for the Ship Mode are quite similar so we can make conclusion on its effect on the profit,as for the segment,Home Office has the highest profit followed by Corporate and Consumer.

Effect of the region on the Profit

```python
fig = plt.figure(figsize=(8,5))
ship=data.groupby('Region').mean().reset_index()
plt.bar(ship['Region'],ship['Profit'],width=0.5)
plt.title('Profit for each Segment')
plt.show()
```

```python
ship
```

There are not much discrepencies region wise other than a lower overall profit in the central region.

```python
data.groupby(['Region','Category']).mean()
```

 If we look at how each sector performs, we can conclude that the central region should improve in the Furniture and Office Supplies sections .

The TOP 10 States in terms of Profit

```python
fig=plt.figure(figsize=(10,8))
state=data.groupby(['State']).mean().reset_index().sort_values(by='Profit',ascending=False).head(10)
plt.bar(state['State'],state['Profit'],width=0.25,color='red')
plt.xticks(rotation=45)
plt.title('Profit for each state')
plt.show()
```

```python
state
```

Effect of each category on the Profit

```python
fig=plt.figure(figsize=(10,8))
category=data.groupby(['Category']).mean().reset_index()
plt.pie(category['Profit'],autopct='%1.2f%%',explode=(0.08,0.04,0.06),colors=('blue','green','red'))
plt.legend(category['Category'])
plt.title('Profit Made in each Category')
plt.show()
```

```python
category
```

Effect of subcategories of the Profit

How much each sub-category is discounted ?

```python
fig=plt.figure(figsize=(10,8))
sub=data.groupby(['Sub-Category']).mean().reset_index().sort_values(by='Discount',ascending=False)
plt.bar(sub['Sub-Category'],sub['Discount'],width=0.3,color='red')
plt.xticks(rotation=45)

plt.show()
```

FURNITURE

```python
figure,axis = plt.subplots(1,2,figsize=(16,8))
category=data.groupby(['Category','Sub-Category']).mean().reset_index()
subcategory=category[category['Category']=='Furniture']
axis[0].bar(subcategory['Sub-Category'],subcategory['Profit'],width=0.2)
axis[0].set_title('Profit made in each sub-category in furniture')
axis[0].axhline(0,color='black')
axis[1].bar(subcategory['Sub-Category'],subcategory['Sales'],width=0.2,color='red')
axis[1].set_title('Sales made in each sub-category in furniture')
plt.show()
```

```python
subcategory.iloc[:,1:]
```

We can see that the biggest loss is in the sub-category of 'tables' but it's also the most sold.So we can maybe minimize this loss by redducing the discount.
